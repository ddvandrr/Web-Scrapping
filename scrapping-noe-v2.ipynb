```python
from selenium import webdriver
from selenium.webdriver.firefox.service import Service
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.firefox import GeckoDriverManager
import pandas as pd
import time

# Setup Firefox headless
options = Options()
options.add_argument("--headless")
driver = webdriver.Firefox(service=Service(GeckoDriverManager().install()), options=options)
wait = WebDriverWait(driver, 10)

url = 'https://silk.menlhk.go.id/index.php/info/exporters'
driver.get(url)

# Tunggu tabel pertama muncul
wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, 'table tbody tr')))
time.sleep(2)

data = []
page = 1
MAX_PAGE = 253

while True:
    print(f"🔄 Ambil data halaman {page}...")

    # Tunggu isi tabel
    rows = driver.find_elements(By.CSS_SELECTOR, 'table tbody tr')
    if not rows:
        break  # Tidak ada baris = akhir halaman

    for row in rows:
        cols = row.find_elements(By.TAG_NAME, 'td')
        if len(cols) >= 3:
            no = cols[0].text.strip()
            nama = cols[1].text.strip()
            alamat = cols[2].text.strip()
            lpvi = cols[3].text.strip()
            slk = cols[4].text.strip()
            masa_berlaku = cols[5].text.strip()
            jenis = cols[6].text.strip()
            data.append([no, nama, alamat, lpvi, slk, masa_berlaku, jenis])

    page += 1
    if page > MAX_PAGE:
        print("⏹️ Batas maksimum halaman tercapai.")
        break

    try:
        # Jalankan fungsi JS doSearch(page)
        driver.execute_script(f"doSearch('{page}')")
        time.sleep(2)
    except Exception as e:
        print(f"❌ Gagal doSearch ke halaman {page}:", e)
        break

# Simpan hasil
df = pd.DataFrame(data, columns=["No", "Nama Eksportir", "Alamat", "Nama LPVI", "No. S-LK", "Masa Berlaku", "Jenis"])
df.to_excel("TMS_SILK_MENLHK.xlsx", index=False)
print("✅ Semua data berhasil disimpan!")
driver.quit()
```
